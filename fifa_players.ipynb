```python
import pandas as pd

df = pd.read_csv('fifa_players.csv')
df.head()
```

```python
serbian_players = df[df['nationality'] == 'Serbia']
serbian_players.head()
```

```python
df.info()
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
mean_value_euro = df['value_euro'].mean()
df['value_euro'] = df['value_euro'].fillna(mean_value_euro)
mean_wage_euro = df['wage_euro'].mean()
df['wage_euro'] = df['wage_euro'].fillna(mean_wage_euro)
mean_release_clause = df['release_clause_euro'].mean()
df['release_clause_euro'] = df['release_clause_euro'].fillna(mean_release_clause)
df['national_rating'] = df['national_rating'].fillna(0)
df['national_jersey_number'] = df['national_jersey_number'].fillna(0)
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
df['national_team'] = df['national_team'].fillna('not-called')
```

```python
df['national_team_position'] = df['national_team_position'].fillna('not-called')
```

```python
df.info()
```

```python
df['is_top_player'] = df['value_euro'] > 40000000
df.info()
```

```python
print(df.columns)

df.info()
```

```python
import matplotlib.pyplot as plt


#distribucija visine igraca

plt.hist(df['height_cm'], bins=30, color='lightgreen', edgecolor='black')
plt.xlabel("Visina igraca (cm)")
plt.ylabel("Broj igraca")
plt.title("Distribucija visine igraca")
plt.show()
```

```python
#distribucija tezine igraca

plt.hist(df['weight_kgs'], bins=30)
plt.xlabel('Tezina igraca (kg)')
plt.ylabel('Broj igraca')
plt.title("Distribucija tezine igraca: ")
plt.show()
```

```python
#Top 10 valuable players

top_10 = df.sort_values(by='value_euro', ascending=False).head(10)

print(top_10[['name', 'value_euro']])
```

```python
#Top 10 drzava po broju igraca (pie chart)

players_by_country = df['nationality'].value_counts()

top_10_countries = players_by_country.head(10)

plt.pie(top_10_countries.values, labels=top_10_countries.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 drzava po broju igraca: ")
plt.axis('equal')
plt.show()
```

```python
#Top 10 drzava sa najskupljim igracima u proseku

avg_value = df.groupby('nationality')['value_euro'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(avg_value.index, avg_value.values, color='gold')
plt.xlabel("Naziv drzave")
plt.xticks(rotation=45)
plt.ylabel("Prosecna vrednost")
plt.title("Top 10 drzava sa najskupljim igracima u proseku: ")
plt.tight_layout()
plt.show()
```

```python
#Prikaži prosečnu vrednost po godinama

avg_value_position = df.groupby('age')['value_euro'].mean()

plt.bar(avg_value_position.index, avg_value_position.values, color='red')
plt.xlabel("Godine")
plt.ylabel("Prosecna vrednost")
plt.tight_layout()
plt.title("Prosecna vrednost igraca po godinama")
plt.show()
```

```python
#Scatter plot visina vs težina po godinama

plt.scatter(df['height_cm'], df['weight_kgs'], c=df['age'], cmap='viridis', alpha=0.7)
plt.xlabel("Visina (cm)")
plt.ylabel("Tezina (kg)")
plt.title("Tezina vs visina po godinama")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Grupisanje godina (npr. 17 do 40)
ages = sorted(df['age'].dropna().unique())
data = [df[df['age'] == age]['wage_euro'].dropna() for age in ages]

plt.figure(figsize=(15, 6))
plt.boxplot(data, labels=ages, patch_artist=True)
plt.xticks(rotation=90)
plt.xlabel('Godine')
plt.ylabel('Plata (EUR)')
plt.title('Boxplot plata po godinama')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Broj igraca visih od 185 cm i vrednijih od 20Miliona

filt = (df['height_cm'] > 185) & (df['value_euro'] > 20000000)

broj_igraca = df[filt].shape[0]

print(f"Broj igrača viših od 185cm i vrednijih od 20M: {broj_igraca}")
```

```python
# Filtriramo igrače mlađe od 22 godine
mladi_igraci = df[df['age'] < 22]

# Grupisanje po poziciji i brojanje
pozicije_broj = mladi_igraci['positions'].value_counts()

# Prikaz prvih 10 najzastupljenijih pozicija
print(pozicije_broj.head(10))
```

```python
#Filtriranje srpskih igraca
srpski_igraci = df[df['nationality'] == 'Serbia']

print(srpski_igraci[['name', 'value_euro', 'age', 'positions']])

svetski_prosek = df['value_euro'].mean(skipna=True).round(2)
srpski_prosek = srpski_igraci['value_euro'].mean(skipna=True).round(2)

print(f"Prosek vrednosti igraca u svetu: {svetski_prosek}, dok je prosek srpskih igraca: {srpski_prosek}")
```

```python
#Top 10 plata po naciji

plata_po_naciji = df.groupby('nationality')['wage_euro'].mean().sort_values(ascending=False).head(10)


plt.bar(plata_po_naciji.index, plata_po_naciji.values, color='yellow')
plt.xlabel("Nacija")
plt.ylabel("Prosecna plata")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#napravljena funkcija koja pokazuje najskupljih 5 igraca po drzavi koju mi napisemo

def top_5_najskupljih_igraca(df, drzava):
    filter = df[df['nationality'] == drzava]
    top5 = filter.sort_values(by='value_euro', ascending=False).head(5)
    return top5[['name', 'nationality', 'value_euro']]

top_5_najskupljih_igraca(df, 'Italy')
```
